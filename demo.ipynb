```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("StudentsPerformance.csv")
print(df.head(10))

print(df.info())

df["Nilai Total"] = df[["math score", "reading score", "writing score"]].mean(axis=1)
df.head(400)

avg_scores = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()

plt.figure(figsize=(10,6))
avg_scores.plot(kind="bar", figsize=(10,6), color=["#799076", "#BBB645", "#9E1803"])

plt.title("Perbandingan Rata-Rata Nilai Tiap Mata Pelajaran Berdasarkan Gender")
plt.xlabel("Gender")
plt.ylabel("Rata-Rata Nilai")
plt.xticks(rotation=0)

plt.legend(["Math", "Reading", "Writing"])
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()

tertinggi = df.loc[df["Nilai Total"].idxmax()]
print("Siswa dengan nilai total tertinggi:")
print(tertinggi)

terendah = df.loc[df["Nilai Total"].idxmin()]
print("\nSiswa dengan nilai total terendah:")
print(terendah)

import matplotlib.pyplot as plt

race_counts = df["race/ethnicity"].value_counts()

custom_colors = ["#FFB3BA", "#BAFFC9", "#BAE1FF", "#FFFFBA", "#E0BBE4"]

max_index = race_counts.idxmax()  
explode_list = [0.1 if race == max_index else 0 for race in race_counts.index]

plt.figure(figsize=(7, 7))
plt.pie(
    race_counts,
    labels=race_counts.index,
    autopct="%1.1f%%",
    startangle=120,
    colors=custom_colors,
    explode=explode_list
)

plt.title("Distribusi Kategori Etnis (Race/Ethnicity)")
plt.show()
```
